```python
x = [0, 1, 2, 3, 4]
y = [1, 1, 0, 0, 2]
n = 5
```

```python
x_mean = sum(x) / n
x_mean
```

```python
x_a = []
x_b = []

for el in x:
    if el <= x_mean:
        x_a.append(el)
    else:
        x_b.append(el)

print(x_a)
print(x_b)

print(sum(x_a) / len(x_a))
print(sum(x_b) / len(x_b))
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
import numpy as np

x = np.array(x)
x = x.reshape(-1, 1)
dtr = DecisionTreeRegressor(max_depth=1)
dtr.fit(x, y)
```

```python
plot_tree(dtr)
```

```python
y_pred = dtr.predict(x)
y_pred
```

```python
y_pred1 = [2.5, 2.5, 0.5, 0.5]
```

```python
from sklearn.metrics import mean_squared_error

y_pred = dtr.predict(x)

mean_squared_error(y, y_pred)
```

```python
mean_squared_error(y, y_pred1)
```
